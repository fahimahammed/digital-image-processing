```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def enhenceBrightness(image, maxIntensity, minIntensity):
    enhencedImage = np.where((image >= minIntensity) & (image <= maxIntensity), np.clip(image + 40, 0, 255), image)
    return enhencedImage
```

```python
originalImage = cv2.imread("../photos/Fractured Spine 746x976.tif", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
# Assume a simplified grayscale image for demonstration (TEST)
image = np.array([[30, 120, 180],
                  [90, 150, 210],
                  [60, 100, 220]])

min_intensity = 100
max_intensity = 230

enhanced_image = np.where((image >= min_intensity) & (image <= max_intensity),
                          np.clip(image + 100, 0, 255),
                          image)

print("Original Image:")
print(image)
print("\nEnhanced Image:")
print(enhanced_image)
```

```python
maxIntensity = 50
minIntensity = 10

enhencedImage = enhenceBrightness(originalImage, maxIntensity, minIntensity)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title(f"original Image")
plt.imshow(originalImage, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Enhenced Image")
plt.imshow(enhencedImage, cmap='gray')
```
