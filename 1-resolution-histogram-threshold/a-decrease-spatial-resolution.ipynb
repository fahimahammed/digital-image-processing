```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
originalImage = cv2.imread("./photos/image5.jpeg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
def decreasedResolution(image): 
    height, width = image.shape
    newHeight = height // 2
    newWidth = width //2
    decreasedImage = np.zeros((newHeight, newWidth))

    for i in range(newHeight):
        for j in range(newWidth):
            decreasedImage[i, j] = np.mean(image[i*2 : ((i*2)+2), j*2 : ((j*2)+2)])
    return decreasedImage
```

```python
decImage = originalImage.copy()
plt.figure(figsize=(16, 8))
for k in range (1, 7):
    plt.subplot(2, 3, k)
    plt.title(f"{decImage.shape}")
    plt.imshow(decImage, cmap='gray')
    decImage = decreasedResolution(decImage)
```
