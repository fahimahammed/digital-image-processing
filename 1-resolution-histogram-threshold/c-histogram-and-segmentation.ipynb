```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
originalImage = cv2.imread("./photos/image3.jpeg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
def generateHistogram(image):
    histogram = np.zeros(256)
    height, width = image.shape

    for i in range(height):
        for j in range(width):
            histogram[image[i, j]] += 1
    return histogram
```

```python
thresholdValue = 128
sigmentedImage = np.where(originalImage > thresholdValue, 255, 0)
sigmentedImage = np.uint8(sigmentedImage)
```

```python
plt.figure(figsize=(12, 12))
grayLevelCount = generateHistogram(originalImage)

#print(grayLevelCount)
plt.subplot(2, 2, 1)
plt.title(f"Histogram (Original Image)")
plt.bar(range(256), grayLevelCount, width=1.0)

plt.subplot(2, 2, 2)
plt.title(f"Original Image")
plt.imshow(originalImage, cmap='gray')

sigmentedImageGrayLevelCount = generateHistogram(sigmentedImage)

plt.subplot(2, 2, 3)
plt.title(f"Histogram (Sigmented Image)")
plt.bar(range(256), sigmentedImageGrayLevelCount, width=1.0)

plt.subplot(2, 2, 4)
plt.title(f"Sigmented Image")
plt.imshow(sigmentedImage, cmap='gray')
```
