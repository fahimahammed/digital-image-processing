```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
originalImage = cv2.imread("../photos/image4.jpeg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
def decreasedResolution(image, numOfBits):
    #decreasedImage = (image // (2 ** (8 - numOfBits))) * (2 ** (8 - numOfBits))
    # Calculate the maximum pixel value for the given bit depth
    maxPixelValue = 2 ** numOfBits - 1
    
    # Reduce the intensity resolution
    decreasedImage = (image * (maxPixelValue / 255)).astype(np.uint8)
    return decreasedImage
```

```python
plt.figure(figsize=(16, 8))

for i in range(1, 9, 1):
    decreasedImage = decreasedResolution(originalImage, 9-i)
    plt.subplot(2, 4, i)
    plt.imshow(decreasedImage, cmap='gray')
    plt.title(f'Bits: {9-i}')
```

Decreasing intensity level resolution in an image results in:

Loss of fine details and reduced image sharpness.
Lower overall contrast.
Increased visibility of quantization artifacts and posterization.
Visual coarseness and pixelation.
Reduced file size due to fewer bits per pixel.
Loss of dynamic range, impacting subtle shading variations.
