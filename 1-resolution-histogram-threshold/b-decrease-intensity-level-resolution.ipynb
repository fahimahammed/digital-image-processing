```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
originalImage = cv2.imread("./photos/image4.jpeg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
def decreasedResolution(image, numOfBits):
    decreasedImage = (image // (2 ** (8 - numOfBits))) * (2 ** (8 - numOfBits))
    return decreasedImage
```

```python
plt.figure(figsize=(16, 8))

for i in range(1, 9, 1):
    decreasedImage = decreasedResolution(originalImage, 9-i)
    plt.subplot(2, 4, i)
    plt.imshow(decreasedImage, cmap='gray')
    plt.title(f'Bits: {9-i}')
```
